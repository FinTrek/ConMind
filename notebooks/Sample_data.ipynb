# Sample datasets
Here I show a sample of the data set released by JD.com. The Raw data is not available in this repository.    
Specifically for this project we require:   
- clickstream history of each user   
- individual products browsed by each user
- order history of users

```python
import pandas as pd
import numpy as np
```

```python
# Loading the user data
users = pd.read_csv('data/JD_user_data.csv')
# Loading the clickstream data
clicks = pd.read_csv('data/JD_click_data.csv')
# Loading the order data
order = pd.read_csv('data/JD_order_data.csv')
```

The clickstream history of a user can be constructed as follows

```python
# Identify a user from the list of orders
i = order.values[0][1]
print(i)
```

```python
# Extract the clickstream data of this user, sorting by time
clicks[clicks['user_ID']==i].sort_values('request_time')
```

```python
# Demographic details of this user 
users[users["user_ID"]==i].T
```

In all above tables, the sku_ID indicates a specific anonymized product browsed by the user.
